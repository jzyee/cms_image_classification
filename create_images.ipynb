```python
import os
import json

cwd = os.getcwd()
json_path = cwd + '/json_files'

with open(json_path + '/Dyjets_json.json') as json_file:
    DYjets = json.load(json_file)

with open(json_path + '/TTjets_json.json') as json_file:
    TTjets = json.load(json_file)

with open(json_path + '/Wjets_json.json') as json_file:
    Wjets = json.load(json_file)

with open(json_path + '/ZZjets_json.json') as json_file:
    ZZjets = json.load(json_file)
```

```python
# Filter the jets, muons and electrons

process.leptonFilter(DYjets, 'muon', minPt = 20, maxPt = False, minEta = False, maxEta = False, tightIDFilter = True, tightIsoFilter = True)
process.leptonFilter(DYjets, 'electron', minPt = 20, maxPt = False, minEta = False, maxEta = False, tightIDFilter = True, tightIsoFilter = True)
process.jetFilter(DYjets, minpt =30, maxpt = False, maxAbsEta = 2.4, jetNumber=False, tightIDFilter = True)

process.leptonFilter(TTjets, 'muon', minPt = 20, maxPt = False, minEta = False, maxEta = False, tightIDFilter = True, tightIsoFilter = True)
process.leptonFilter(TTjets, 'electron', minPt = 20, maxPt = False, minEta = False, maxEta = False, tightIDFilter = True, tightIsoFilter = True)
process.jetFilter(TTjets, minpt =30, maxpt = False, maxAbsEta = 2.4, jetNumber=False, tightIDFilter = True)

process.leptonFilter(Wjets, 'muon', minPt = 20, maxPt = False, minEta = False, maxEta = False, tightIDFilter = True, tightIsoFilter = True)
process.leptonFilter(Wjets, 'electron', minPt = 20, maxPt = False, minEta = False, maxEta = False, tightIDFilter = True, tightIsoFilter = True)
process.jetFilter(Wjets, minpt =30, maxpt = False, maxAbsEta = 2.4, jetNumber=False, tightIDFilter = True)
```

```python
im.generateAllhistograms(DYjets, DirName = 'DY After Lepton and Jet Filter', origen = 'DY')
im.generateAllhistograms(TTjets, DirName = 'TTjets After Lepton and Jet Filter', origen = 'TTjets')
im.generateAllhistograms(Wjets, DirName = 'Wjets After Lepton and Jet Filter', origen = 'Wjets')
```

```python
DY_Class = training.selectClass_KnownClass(DYjets, ClassName = 'DYjets', ClassNumber = '0')
TT_Class = training.selectClass_KnownClass(TTjets, ClassName = 'TTjets', ClassNumber = '1')
W_Class = training.selectClass_KnownClass(Wjets, ClassName = 'Wjets', ClassNumber = '2')
```

```python
test_DY_Class = []
test_TT_Class = []
test_W_Class = []
test_ZZ_Class = []
```

```python
print(len(DY_Class))
print(len(TT_Class))
print(len(W_Class))
```

```python
for i in range(0, 800):
    test_DY_Class.append(DY_Class.pop())
    test_W_Class.append(W_Class.pop())
    test_TT_Class.append(TT_Class.pop())
    test_ZZ_Class.append(ZZ_Class.pop())   
```

```python
test_Class = test_DY_Class +  test_W_Class + test_TT_Class + test_ZZ_Class
random.shuffle(test_Class)
len(test_Class)
```

```python
val_ZZ_Class = []
val_DY_Class = []
val_W_Class = []
val_TT_Class = []

for i in range(0,800):
    val_DY_Class.append(DY_Class.pop())
    val_W_Class.append(W_Class.pop())
    val_TT_Class.append(TT_Class.pop())
    val_ZZ_Class.append(ZZ_Class.pop())
    
val_Class = val_DY_Class +  val_W_Class + val_TT_Class + val_ZZ_Class
random.shuffle(val_Class)
len(val_Class)
```

```python
train_ZZ_Class = []
train_DY_Class = []
train_W_Class = []
train_TT_Class = []

for i in range(0,2400):
    train_DY_Class.append(DY_Class.pop())
    train_W_Class.append(W_Class.pop())
    train_TT_Class.append(TT_Class.pop())
    train_ZZ_Class.append(ZZ_Class.pop())
    
train_Class = train_DY_Class +  train_W_Class + train_TT_Class + train_ZZ_Class
random.shuffle(train_Class)
len(train_Class)
```

```python
# Delete the data stored in test.txt train.txt and val.txt
test=open('Training Files/test.txt','w')
test.close()
val=open('Training Files/val.txt','w')
val.close()
train=open('Training Files/train.txt','w')
train.close() 
```

```python
# Write in text training files
# Writes the txt files in Training Files folder
for n in range(0, len(test_Class)):
    training.writeTest(test_Class[n])
for n in range(0, len(val_Class)):
    training.writeVal(val_Class[n])
for n in range(0, len(train_Class)):
    training.writeTrain(train_Class[n])
```
